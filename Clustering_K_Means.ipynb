```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.cluster import KMeans
from sklearn.externals import joblib
```

```python
kmeans_data_df = pd.read_csv("C:\\ADS\\LendingClubData\\loan_data\\clustering_data.csv",low_memory=False)
```

```python
kmeans_data_df = kmeans_data_df.drop(['grade'], axis=1)
```

```python
kmeans_data= kmeans_data_df.drop(['int_rate'], axis=1)
```

```python
normalized_data = preprocessing.normalize(kmeans_data)
```

```python
clusters = [4,6,8,10,12]
def cluster_data(k, dataset) :
    kmeans = KMeans(n_clusters=k, random_state=0, verbose=1)
    kmeans.fit(dataset)
    return kmeans.inertia_
inertia_values  = [cluster_data(k, normalized_data) for k in clusters]
inertia_values
```

```python
elbow_curve = pd.DataFrame()
elbow_curve['k'] = clusters
elbow_curve['inertia'] = inertia_values
elbow_curve.plot(kind='line', x='k', y='inertia',marker = 'o')
```

```python
kmeans = KMeans(n_clusters=8, random_state=0)
kmeans.fit(normalized_data)
```

```python
joblib.dump(kmeans,'C:\\ADS\\LendingClubData\\loan_data\\Kmeans_model.pkl')
```

```python
cluster_df = pd.DataFrame(kmeans.labels_,columns=['cluster'])
```

```python
cluster_df.shape
```

```python
data_df = pd.concat([kmeans_data_df,cluster_df],axis=1)
```

```python
clustered_df = data_df.groupby(['cluster'])
clustered_df = list(clustered_df)
```

```python
clustered_df
```

```python
count = 0
kmeans_cluster = []
for cluster in clustered_df:
    kmeans_cluster_df = clustered_df[count][1]
    kmeans_cluster.append(kmeans_cluster_df)
    print("Cluster", count, "has", len(kmeans_cluster_df), " rows")
    csv_name = "C:\\ADS\\LendingClubData\\loan_data\\Kmeans_Cluster_" + str(count) + ".csv"
    kmeans_cluster_df.drop('cluster', 1, inplace=True)
    kmeans_cluster_df.to_csv(csv_name, index = False)
    count += 1
```
